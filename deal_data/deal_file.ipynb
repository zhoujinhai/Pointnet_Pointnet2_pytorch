### 一、从文件夹中挑选出模型stl文件和对应的牙龈线点pts文件

```python
import os 
from shutil import copyfile, move
```

```python
def find_all_file(base):
    for root, ds, fs in os.walk(base):
        for f in fs:
            fullname = os.path.join(root, f)
            yield fullname
```

```python
save_path = "/home/heygears/work/MeshCNN/datasets/tooth"
stl_path = os.path.join(save_path, "stl")
pts_path = os.path.join(save_path, "pts")

if not os.path.isdir(stl_path): 
    os.makedirs(stl_path)
if not os.path.isdir(pts_path):
    os.makedirs(pts_path)

i = 1
j = 1
for file_path in find_all_file("/media/heygears/资料/data/dentalModel"):   # os.getcwd()):
    basename = os.path.basename(file_path)
    dirs = file_path.split("/")
    sub_path = dirs[-2]
    if file_path.find(".stl") != -1:
        print(i, " copy ", file_path)
        i += 1
        copyfile(file_path, os.path.join(stl_path, sub_path + "_" + basename))
    if file_path.find(".pts") != -1:
        print(j, " copy ", file_path)
        copyfile(file_path, os.path.join(pts_path, sub_path+ "_" + basename))
        j += 1
```

### 二、预处理stl文件

```python
import sys
sys.path.append(r"/home/heygears/work/package/hgapi")
import hgapi

import os 
import glob
```

```python
def healing(mesh_path, save_path):
    mesh_healing = hgapi.MeshHealing()
    mesh = hgapi.Mesh()
    stl_handler = hgapi.STLFileHandler()
    stl_handler.Read(mesh_path, mesh)
    mesh_healing.RemoveSmallerIslands(mesh) # 删除孤立块
    mesh_healing.RemoveDanglingFace(mesh)   # 删除悬边
    stl_handler.Write(save_path, mesh)
```

```python
# 去掉夹治具
model_dir = "/home/heygears/work/Tooth_data_prepare/tooth/stl"
save_dir = "/home/heygears/work/Tooth_data_prepare/tooth/healing_stl/"
model_list = glob.glob(os.path.join(model_dir, "*.stl"))
print(len(model_list))
for i, model_path in enumerate(model_list):
    model_name = os.path.basename(model_path)
    save_path = os.path.join(save_dir, model_name)
    healing(model_path, save_path)
    print("{} {} is healed".format(i, model_path))
```

### 三、将模型有stl格式转为obj格式

```python
import vtk
from vedo import load, write, Points
import glob
import os
from multiprocessing import Process
from tqdm import tqdm
```

```python
# write 默认只保留5位，这里根据源码编写转换函数 https://vedo.embl.es/_modules/vedo/io.html#write
def stl_2_obj(objct, file_out_path):
    """
    stl model convert to obj model
    @objct: stl model
    @file_out_path: obj保存地址
    """

    fr = file_out_path.lower()
    
    assert fr.endswith(".obj"), "fileoutput 需以obj为后缀"
    outF = open(file_out_path, "w")
    outF.write('# OBJ file format with ext .obj\n')
    outF.write('# File generated by vedo\n')

    for p in objct.points():
        # outF.write("v {:.5g} {:.5g} {:.5g}\n".format(*p))
        outF.write("v {:.10f} {:.10f} {:.10f}\n".format(*p))

    for i,f in enumerate(objct.faces()):
        fs = ''
        for fi in f:
            fs += " {:d}".format(fi+1)
        outF.write('f' + fs + '\n')

    for l in objct.lines():
        ls = ''
        for li in l:
            ls += str(li+1)+" "
        outF.write('l '+ ls + '\n')

    outF.close()
    return objct
    
```

```python
def stl_2_obj_batch(stl_list, obj_save_path):
    """
    stl格式批量转换为obj格式
    """
    for i in tqdm(range(len(stl_list))):
        stl_path = stl_list[i]
        save_name = os.path.basename(stl_path).replace(".stl", ".obj")
        obj_path = os.path.join(obj_save_path, save_name)
        stl_model = load(stl_path)
        stl_2_obj(stl_model, obj_path)     
        # print("The {} model is converted!".format(i))
```

```python
def parallel_stl_2_obj(stl_list, obj_save_path, n_workers=8):
    """
    多进程处理
    """
    if len(stl_list) < n_workers:
        n_workers = len(stl_list)
    chunk_len = len(stl_list) // n_workers

    chunk_lists = [stl_list[i:i+chunk_len] for i in range(0, (n_workers-1)*chunk_len, chunk_len)]
    chunk_lists.append(stl_list[(n_workers - 1)*chunk_len:])
    
    process_list = [Process(target=stl_2_obj_batch, args=(chunk_list, obj_save_path)) for chunk_list in chunk_lists]
    for process in process_list:
        process.start()
    for process in process_list:
        process.join()
```

```python
stl_model_path = "/home/heygears/work/Tooth_data_prepare/tooth/healing_stl"
obj_model_path = "/home/heygears/work/Tooth_data_prepare/tooth/obj"

stl_list = glob.glob(os.path.join(stl_model_path, "*.stl"))
print(len(stl_list))

parallel_stl_2_obj(stl_list, obj_model_path, 8)
```
